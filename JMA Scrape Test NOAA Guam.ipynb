```python
from bs4 import BeautifulSoup
import requests
```

```python
#url = "https://www.jma.go.jp/bosai/amedas/#amdno=44132&area_type=offices&area_code=130000&format=table1h&elems=53614"
url = "https://www.ndbc.noaa.gov/station_page.php?station=aprp7&uom=M&tz=JST"
# Barometer elevation: 5.1m
```

```python
page = requests.get(url)
soup = BeautifulSoup(page.content, "html.parser")
print(soup.prettify())
```

```python
table = soup.find(id="data")
print(table)
```

```python
tr = table.find_all("tr")
```

```python
td = tr[4].find_all("td")
td
```

```python
text = td[2].get_text()
pressure = text[:-3]
print(pressure)
```
